```python
#변수 분포 탐색 
import matplotlib.pyplot as plt
numerical_columns = ['play_free_score', 'change_score',
       'market_score', 'money_score', 'lotate_score', 'preschol_score',
       'park_score', 'schol_score', 'class_score', 'hopital_score1',
       'hopital_score2']

fig  = plt.figure(figsize = (16,20))
ax = fig.gca()
score[numerical_columns].hist(ax=ax)
plt.show()
```

```python
#변수들 간의 상관관계 탐색 

cols = ['play_free_score', 'change_score',
       'market_score', 'money_score', 'lotate_score', 'preschol_score',
       'park_score', 'schol_score', 'class_score', 'hopital_score1',
       'hopital_score2']

corr = score[cols].corr(method = 'pearson')
```

```python
corr
```

```python
#상관관계 시각화로 나타내기
fig = plt.figure(figsize = (16, 12))
ax = fig.gca()

sns.set(font_scale=1.5)
hm = sns.heatmap(corr ,
                 annot=True, 
                 fmt='.2f',
                 annot_kws={'size': 15},
                 yticklabels=cols,
                 xticklabels=cols,
                 ax=ax)
plt.tight_layout()
plt.show()
```
