```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
import pandas as pd 

# Chrome의 경우 chromedriver의 위치를 지정해준다.
# driver = webdriver.Chrome('E:/Ella/RStudio/DataAnalysis/프로젝트/chromedriver.exe')
driver = webdriver.Chrome('C:/Users/tjoeun-jg-303/Desktop/chromedriver.exe')

# 암묵적으로 웹 자원 로드를 위해 3초까지 기다려 준다.
#driver.implicitly_wait(3)
time.sleep(3)

# selenium으로 url에 접근
driver.get('https://www.cpf.go.kr/front/sub01/sub0101.do#')
```

```python
#함수 만들기 
def kid_address(string):
    driver.find_element_by_id('areaCode1').send_keys('서울')
    time.sleep(1)
    driver.find_element_by_id('areaCode2').send_keys('%s' %string)
    time.sleep(1)

    #검색하기 누르기 
    #<a href = '#', onclick = 기타명령어> 테스트2 </a> 경우 (fullxpath를 이용해서 클릭하기)
    # /html/body/div[2]/div[2]/div[2]/form/div/div/table/tbody/tr[1]/td[7]/a
    driver.find_element_by_xpath("//div[2]/div[2]/div[2]/form/div/div/table/tbody/tr[1]/td[7]/a").click()
    time.sleep(2)
    
    i = 1
    kbdata1 = []
    while i == 1:
        i +=1
        driver.find_element_by_xpath('//*[@id="board"]/div/div/a[2]').click()
        time.sleep(3)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        time.sleep(2)

        tdata = soup.find('div', {'id':'board'})
        series = tdata.find('tbody')
        kbdata_spec = series.text
        #     kbdata.append(kbdata_spec)
        table_rows = tdata.find_all('tr')

        #res 라는 list에 row별로 담아둔다. 
        res = []
        for tr in table_rows:
            td = tr.find_all('td')
            row = [tr.text.strip() for tr in td if tr.text.strip()]
            if row:
                res.append(row)
        kbdata1.append(res)

#두번째부터 마지막 페이지까지 크롤링

    i = 1
    kbdata2=[]
    while i <= 10 :
        try:
            i = i + 1

            for page in range(3,13):

                driver.find_element_by_xpath('//*[@id="board"]/div/div/a[%s]' %page).click()
                time.sleep(3)
                # 페이지 소스 가져오기
                html = driver.page_source
                soup = BeautifulSoup(html, 'html.parser')
                time.sleep(1)
                tdata = soup.find('div', {'id':'board'})
                series = tdata.find('tbody')
                kbdata_spec = series.text
            #     kbdata.append(kbdata_spec)
                table_rows = tdata.find_all('tr')

                #res 라는 list에 row별로 담아둔다. 
                res = []
                for tr in table_rows:
                    td = tr.find_all('td')
                    row = [tr.text.strip() for tr in td if tr.text.strip()]
        #         if row:
                    res.append(row)


                kbdata2.append(res)

                if len(res) <11 :
                    break;
        except:
            print('코드업로드 완성')
                
    kbdata = kbdata1 + kbdata2

    #DateFrame을 만들기 위한 리스트 정리
    data_all = []
    for i in kbdata:
        for g in i :
            data_all.append(g)
            

    #None 값 fillter로 삭제하기 
    data_all = list(filter(None, data_all))
    
    #중복 제거 
    new_data = []
    for data in data_all:
        if data not in new_data:
            new_data.append(data)

    #표로 만들기
    df = pd.DataFrame(new_data, columns = ["number", "place", "type","address"])       

    #저장하기
    df.to_csv('D:/Ella/project/어린이집/kid_address_%s.csv'%(string))
    df.info()
```
